# Spark Hello World

```python
from pyspark.sql import SparkSession
```

## Spark 4.0.1 with Comet

```python
spark_comet = SparkSession.builder \    .appName("SparkCometHelloWorld") \    .master("spark://spark-master-4.0.1-comet:7077") \    .getOrCreate()df_comet = spark_comet.createDataFrame([("hello", "comet")], ["col1", "col2"])df_comet.show()print(f"Spark version: {spark_comet.version}")spark_comet.stop()
```
